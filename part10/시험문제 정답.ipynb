### 제공되는 pokemon.csv를 사용하여 ‘type_1’을 x축으로 하여, 각각 몇 개씩 있는지를 bar chart로 시각화 한다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon_df = pd.read_csv('pokemon.csv')
```

```python
pokemon_df.head()
```

```python
sns.countplot(data = pokemon_df, x = 'type_1')
plt.xticks(rotation = 45)
plt.savefig('pokemon.png')
```

### 제공되는 pokemon.csv를 사용하여 speed의 단위를 5씩 끊어서, 그에 따른 히스토그램을 나타내시오

```python
pokemon_df.head()
```

```python
bins = np.arange(0, pokemon_df['speed'].max(), 5 )
```

```python
plt.hist(data = pokemon_df, x = 'speed', bins= bins, rwidth=0.7)
plt.savefig('pokemon2.png')
```

### fuel_econ.csv 파일을 이용해서 엔진 크기(displ) 와 이산화탄소 배출량(co2)의 관계를 히트맵으로 나타내세요.

```python
fuel_econ_df = pd.read_csv('fuel_econ.csv')
```

```python
fuel_econ_df.head()
```

```python
plt.hist2d(data = fuel_econ_df,  x = 'displ', y ='co2', cmin=0.5)
plt.colorbar()
plt.savefig('fuel_econ.png')
```
