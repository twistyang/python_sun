```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('pokemon.csv')
```

```python
df.head()
```

```python
plt.title('Type1 Count Plot')
sb.countplot(data=df, x='type_1')
plt.xticks(rotation = 90)
plt.savefig('type.png')
plt.show()
```
