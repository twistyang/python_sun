# WORD CLOUD 사용하여, 유저들이 어떤 단어를 많이 사용하였는지 시각화 해봅니다. 

# 먼저 WORD CLOUD 를 이용하기 위해서, verified_reviews 를 하나의 문자열로 만들겠습니다.

## 실습 1. verified_reviews 를 하나의 리스트로 만듭니다.

```python
df_alexa.shape
```

```python
word_list = list(df_alexa['verified_reviews'].values)
```

## 실습 2. 위의 words 리스트를, " " (공백) 으로 합쳐서, 하나의 문자열로 만듭니다.

```python
word_as_one_string = " ".join(word_list)
```

```python
word_as_one_string
```

## 실습 3. WordCloud 를 이용하여, 화면에 많이 나온 단어들을 시각화 합니다.

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(background_color='white')
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wc.generate(word_as_one_string))
plt.axis('off')
plt.title("Reviews")
plt.show()
```

## 실습 4. Data Cleaning 과 Feature Engineering

```python
from sklearn.feature_extraction.text import CountVectorizer
```
